```python
# Summary 
# Supervised learning is a type of machine learning where the model is trained on a labeled dataset, predicting the outcome of new unseen data. 
# It is used for regression and classification problems.

# Unsupservised learning is a type of machine learning where the model is trained on an unlabeled dataset, finding patterns and relationships in the data.
# It is used for clustering and association problems,dimensionality reduction.

# Definition
# Supervised Learning : Learns from labelled data to predict outputs for new inputs
# Unsupervised Learning : Learns from unlabelled data to find hidden patterns or intrinsic structures in input data.
```

```python
import pandas as pd
data = pd.read_csv('heart.csv')

data.info()
```

```python
# checking for empty cells
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
print(data.to_string())
```

```python
data.columns
```

```python
data.describe().T
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(data['age'],color='blue',bins=10,kde=True)
plt.title("Age Distribution")
plt.show()

print(data.to_string())
```

```python
sns.boxplot(y="thalach", x='oldpeak', data=data, palette='plasma_r')
plt.xticks(rotation=90)
plt.title('Thalach against old peak')
plt.show()
```

```python
print(data.head(20).to_string())
```

```python
# correlation between age and cp
corr = data[['age','cp','thalach']].corr()
sns.heatmap(corr,cmap='RdBu',annot=True ,vmax=1,vmin=-1)
plt.title('Correlation between Age , Cp and thalach')
plt.show()
```

```python
sns.histplot(data['thalach'],color='red',bins=10,kde=True)
plt.title('Thalach Over view')
plt.show()
```

```python
corr = data[['age','cp']].corr()
sns.heatmap(corr,cmap='RdBu',annot=True ,vmax=1,vmin=-1)
plt.title('Correlation between Age and Cp')
plt.show()
```
